```python
import yfinance as yf
import pandas as pd
```

```python
symbols_and_weight = {
    "AAPL": 0.1,
    "GOOGL": 0.1,
    "MSFT": 0.1,
    "JNJ": 0.1,
    "KO": 0.1,
    "^IRX": 0.5,
}
```

```python
start_date = "2022-01-01"
end_date = "2023-12-31"
```

```python
asset_data={}
```

```python
for symbol,weight in symbols_and_weight.items():
    asset_data[symbol] = yf.download(symbol,start = start_date, end = end_date)
    asset_data[symbol]['Weight']= weight
    
```

```python
asset_data
```

```python
asset_data["AAPL"]
```

```python
for symbol in symbols_and_weight.keys():
    asset_data[symbol]['Return'] = asset_data[symbol]['Close'].pct_change()
    asset_data[symbol]['Volatility'] = asset_data[symbol]['Return'].rolling(window=20).std()
    asset_data[symbol]['MovingAverage'] = asset_data[symbol]['Close'].rolling(window=10).mean()
```

```python
asset_data
```

```python
all_data = pd.concat(list(asset_data.values()),axis =1, keys = list(asset_data.keys()))
all_data
```

```python
all_data_cleaned = all_data.dropna()
```

```python
all_data_cleaned
```

```python
all_data_cleaned.to_csv('process.csv',index = True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score
models={}
predictions = {}
stock_names = all_data_cleaned.columns.get_level_values(0).unique()
```

```python
for stock in stock_names:
    X = all_data_cleaned[stock][['MovingAverage','Volatility']].values
    y = all_data_cleaned[stock]['Close'].values
    model = LinearRegression()
    model.fit(X,y)
    last_day = all_data_cleaned[stock][['MovingAverage','Volatility']].iloc[-1].values.reshape(1,-1)
    next_day_close_predicted = model.predict(last_day)
    models[stock] = model
    r2 = r2_score(y,model.predict(X))
    predictions[stock] = next_day_close_predicted
    print(f'predict{stock} prices:{next_day_close_predicted}')
    print(f'R2:{r2}')
```
